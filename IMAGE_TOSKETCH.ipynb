```python
import cv2
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("C:/Users/Naina Sahu/OneDrive/New folder/car.jpeg")
```

```python
plt.imshow(img)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.max()
```

```python
img.min()
```

```python
inv=255-img
```

```python
plt.imshow(inv, cmap='gray')
```

```python
blur=cv2.GaussianBlur(inv, ksize=(21,21), sigmaX=0, sigmaY=0)
```

```python
plt.imshow(blur, cmap ='gray')
```

```python
dodge= lambda image, mask: cv2.divide(image, 255-mask, scale=256)
```

```python
blende=dodge(img, blur)
```

```python
plt.imshow(blende, cmap='gray')
```

```python
cv2.imwrite('Car.jpg',blende)
```
